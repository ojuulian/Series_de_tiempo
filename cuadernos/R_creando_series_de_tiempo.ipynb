```r
#----------------------------
## Importación de base de datos con varias variables
#----------------------------
library(readr)
library(readxl)
Datos2 <- read_excel("C:\\Users\\YULY\\Desktop\\Desktop\\UNAL 2018\\series de tiempo\\codigos\\Base_Accidentes.xlsx")
#View(Datos2)
head(Datos2)
```

```r
#----------------------------
## Series con frecuencia mensual
#----------------------------
HER=ts(Datos2$HER,start=c(2005,1),end=c(2018,7),frequency=12)
IC=ts(Datos2$IC,start=c(2005,1),end=c(2018,7),frequency=12)
ISE=ts(Datos2$ISE,start=c(2005,1),end=c(2018,7),frequency=12)
```

```r
HER
```

```r
#----------------------------
# Graficos
#----------------------------
plot(HER, main="Número de heridos")
```

```r
plot(IC, main="Indice IC")
plot(ISE, main="Indice ISE")
```

```r
#----------------------------
## Graficar varias series al tiempo
#----------------------------
plot(ISE, main="",ylab="",xlab="Fecha", ylim=c(60,300))
lines(IC, main="", col="red",ylab="IC",xlab="Fecha")
```

```r
par(mfrow=c(2,1))
plot(ISE, main="",ylab="ISE",xlab="Fecha")
plot(IC, main="", col="red",ylab="IC",xlab="Fecha")
```

```r
#----------------------------
## Estad?sticas descriptivas
#----------------------------
summary(HER)
summary(Datos2[,c(3:21)])
```

```r
###########################################
### Series de tiempo  (uso de funcion "ts")
### creando objeto de serie de tiempo.
###########################################

#rm(list=ls())
#----------------------------
## Importación de base de datos COLCAP
#----------------------------
library(readr)
library(readxl)
Colcap <- read_excel("C:\\Users\\YULY\\Desktop\\Desktop\\UNAL 2018\\series de tiempo\\codigos\\Colcap.xlsx", 
                     sheet = "Colcap", col_types = c("date","numeric"))
```

```r
fechas=as.Date(Colcap$Fecha)
library(xts)
colcap=xts(Colcap$ValorCOLCAP,order.by = fechas)
plot(colcap)
```

```r
########
retornos=diff(log(colcap))[2:length(colcap)]
head(retornos)
plot(retornos)
```
