Creando una simulaci√≥n de una caminata aleatoria a partir de $T=200$ numeros aleatorios tales que $x\sim N(0,1)$ 

```r
#ls()
#rm(list=ls())
```

```r
set.seed(20)
x<-rnorm(200, mean = 0, sd = 1)

s<-rep(0,length(x))
for(i in 1:length(x)){
    if(i>1){
      s[i]<-sum(x[1:i])
    }
  else{
    s[i]==0
  }
  }

contador<-seq(0,199)
```

```r
plot(contador,s,main = "Caminata Aleatoria",pch=20,type = "o")
#plot(contador,s,main = "Caminata Aleatoria",pch=20,type = "l", xlab="Tiempo", ylab="S_n")
plot(contador,s,main = "Caminata Aleatoria",pch=20,type = "l", xlab="Tiempo", ylab=expression(S_t))
```
