Si $\{X_t \}$ es un proceso estocástico tal que $E[X_t]=0$ , $E[X_t^2]=\sigma^2$ y que para cada n las variables $X_{t_1}, X_{t_2},...,X_{t_n} $ son independientes y tienen la misma distribución, entonces el proceso   $\{X_t \}$ es llamado IID y se denota:

$$ \{X_t \} \sim IID(\mu , \sigma)$$

Se procede a hacer la simulacion de 200 valores aleatorios de una distribucion normal con media $\mu$ y varianza $\sigma^2$, En consecuencia, se usa la funcion numpy para generar $n$ numeros aleatorios de una distribucion normal y se realiza su respectiva representación gráfica:

```python
import numpy as np     
import pandas as pd
import math 
import matplotlib.pylab as plt #libreria para graficar
from statsmodels.tsa.stattools import acf 
from statsmodels.graphics.tsaplots import plot_acf
```

```python
n=200 #Tamaño de la serie
l=50 
sigma=2

####Simulación IID####
serieIID=pd.Series(np.random.normal(0, sigma, n))
plt.plot(serieIID)
```

```python
acf(serieIID,nlags=l,unbiased=False) 
plot_acf(serieIID,lags=l,unbiased=False)  
```
