### Simulando AR(1)

```r
#rm(list=ls())
#ls()
```

```r
#----------------
# Simulando AR(1)
#----------------
Z<-rnorm(100,0,1)
phi<-0.2
n<-length(Z)

X<-c(rep(0,n))
X[1]<-Z[1]

for (i in 2:n) {
  X[i]<-X[i-1]*phi+Z[i]
}
AR1<-ts(X)
plot(AR1)
acf(AR1)
```

```r
#Covarianza(0)
1/(1-phi^2)  

# media muestral
med<-sum(AR1)/n ; med
```

```r
# autocovarianza muestral para la media del proceso
h<-50

var_med<-0
for (i in 1:h) {
  var_med<-var_med+(1-h/n)*phi^h
}
var_med

# Autocovarianza estimada
h<-20
n<-length(AR1)
sup<-h+1
sum<-c(rep(0,sup))
for (i in 1:sup) {
  su<-n-h
  for (t in 1:su) {
    sum[i]<-sum[i]+(AR1[t+i-1]-med)*(AR1[t]-med)
  }
}
acvf<-sum/n
##
acrf<-acvf/acvf[1]
plot(acvf,type = "h")
abline(a=0,b=0)

plot(acrf,type = "h")
abline(a=0,b=0)
```

### Simulando MA(1)

```r
#----------------
# Simulando MA(1)
#----------------
n<-100
theta<-0.7
Z<-rnorm(n,0,1)
for (i in 2:n) {
  Z[i]<-Z[i]+theta*Z[i-1]
}
MA1<-Z
plot(ts(MA1))
acf(MA1)

h<-20
n<-length(MA1)
sup<-h+1
sum<-c(rep(0,sup))

for (i in 1:sup) {
  su<-n-h
  for (t in 1:su) {
    sum[i]<-sum[i]+(MA1[t+i-1]-med)*(MA1[t]-med)
  }
}
acvf<-sum/n
```

```r
plot(acvf,type="h")
abline(0,0)
acf(MA1)
```
